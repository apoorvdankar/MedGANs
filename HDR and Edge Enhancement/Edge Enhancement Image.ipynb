```python
from PIL import Image
from PIL import ImageFilter
```

```python
# Open an already existing image

imageObject = Image.open("./result.jpg")
```

```python
# Apply edge enhancement filter

edgeEnahnced = imageObject.filter(ImageFilter.EDGE_ENHANCE)
```

```python
# Apply increased edge enhancement filter

moreEdgeEnahnced = imageObject.filter(ImageFilter.EDGE_ENHANCE_MORE)
```

```python
edgeEnahnced.save('edge_enhanced.jpg')
moreEdgeEnahnced.save('more_edge_enhanced.jpg')
```

```python
import skimage.io as io
```

```python
import cv2

normal_image = cv2.imread('test.png')
hdr_image = io.imread('result.jpg')

edge_enhanced = io.imread('edge_enhanced.jpg')
more_edge_enhanced = io.imread('more_edge_enhanced.jpg')
```
